### 反向传播

* 线性函数： $z= Wx + b $
* 线性函数被激活后： $ a = \sigma (z)$
* 目标函数：$ minimize  J = max(\Delta + s_c -s,0) $   先让$ \Delta =1 $

* $ w_{jk}^l $ : 第l层第k个神经元到第l+1层的第j个神经元的权重

* $ w_{jk}^l $ 的变动 $ \Delta w_{jk}^l$ 对于目标函数的影响 是 $ \Delta C $
    * $ \Delta C = {  \partial C \over   \partial w_{jk}^l } \Delta w_{jk}^l $
    * 到目标函数的一条线路影响：$ \Delta C = {  \partial C \over   \partial a_m^L } {  \partial a_m^L \over   \partial a_m^{L-1} } ... {  \partial a_j^{l} \over   \partial w_{jk}^{l} } \Delta w_{jk}^l $
    * 所有线路的影响： $ \Delta C = \sum{  \partial C \over   \partial a_m^L } {  \partial a_m^L \over   \partial a_m^{L-1} } ... {  \partial a_j^{l} \over   \partial w_{jk}^{l} } \Delta w_{jk}^l $
    
* $ \delta_j^k $ 是 第k层第j个神经元的误差 $ \delta_j^k = { \partial J \over \partial z_j^k } $
* 反向传播算法：
    * 误差：$ \delta_j^l = \nabla_a 点积   \sigma(z^l)$ 
    * 误差与下一层误差的关系：$ \delta_j^l = (W^{l+1})^T \delta_j^{l+1} 点积 \sigma(z^l) $
    * 偏置项和误差的关系： $ { \partial C\over\partial b^l } = \delta_j^l   $
    * 权重和误差的关系： $ { \partial C\over\partial w_{jk}^l } =  \alpha_k^{l-1} \delta_j^l   $
